# Task 2: Data Pre Processing

```python
Main["CRIM"].fillna(Main["CRIM"].mean(), inplace=True) # filling CRIM with mean value as Crime rate matters most of the time
Main["ZN"].fillna(0, inplace=True) # fill ZN with mode value, because 360 of the entries are 0 therefore making the value 0
Main["INDUS"].fillna(Main["INDUS"].mean(), inplace=True) # fill INDUS with mean value,because as non retail business can 
                                                         #be causing more material gain
Main["CHAS"].fillna(0, inplace=True) # fill CHAS with mode value, because 452 of the entries are 0
Main["AGE"].fillna(Main["AGE"].mean(), inplace=True) # fill AGE with mean value, this is crucial factor as per lenders
Main["LSTAT"].fillna(Main["LSTAT"].mean(), inplace=True) # fill LSTAT with mean value, as lower status may 
                                                         #drags property value down.
```

```python
Main.isnull().sum() # Checking for any missing values again.
```

```python
outliers = Main.quantile(.97) # dealing with the outliers seen in the boxplots above
Main = Main[(Main['CRIM']<outliers['CRIM'])]
Main = Main[(Main['ZN']<outliers['ZN'])]
Main = Main[(Main['RM']<outliers['RM'])]
Main = Main[(Main['DIS']<outliers['DIS'])]
Main = Main[(Main['PTRATIO']<outliers['PTRATIO'])]
Main = Main[(Main['B']<outliers['B'])]
Main = Main[(Main['LSTAT']<outliers['LSTAT'])]
```

# Task 3: Machine Learning [Model Building]

```python
# Function to call all Regression Algorithms
def regression_funct(x,y):
    #importing Regression libraries
    from sklearn.linear_model import LinearRegression
    from sklearn.linear_model import Ridge,Lasso
    from sklearn.linear_model import ElasticNet
    from sklearn.tree import ExtraTreeRegressor
    from sklearn.ensemble import GradientBoostingRegressor
    from sklearn.neighbors import KNeighborsRegressor
    from xgboost import XGBRegressor
    
    from sklearn.model_selection import train_test_split
    from sklearn.metrics import mean_squared_error,r2_score,mean_absolute_error
    
    # split the data in Train and Test 
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)

    L=LinearRegression()
    R=Ridge()
    Lass=Lasso()
    E=ElasticNet()
    ExTree=ExtraTreeRegressor()
    GBR=GradientBoostingRegressor()
    KN=KNeighborsRegressor()
    XGBC= XGBRegressor()
    
    algos=[L,R,Lass,E,ExTree,GBR,KN,XGBC]
    algo_names=['LinearRegression','Ridge','Lasso','ElasticNet','ExtraTreeRegressor','GradientBoostingRegressor','KNeighborsRegressor','XGBRegressor']
    r_squared=[]
    rmse=[]
    mae=[]
    
    result=pd.DataFrame(columns=['R_Squared','RMSE','MAE'],index=algo_names) # create df with results
    
    for item in algos: # fit and predict model with all algos and append the results in their lists
        item.fit(x_train,y_train)
        item.predict(x_test)
        r_squared.append(r2_score(y_test,item.predict(x_test)))
        rmse.append((mean_squared_error(y_test,item.predict(x_test)))**.5)
        mae.append(mean_absolute_error(y_test,item.predict(x_test)))
        
    result.R_Squared=r_squared
    result.RMSE=rmse
    result.MAE=mae
        
    return result.sort_values('R_Squared',ascending=False)
```

```python
x,y=Main.drop("MEDV",axis=1),Main[["MEDV"]] # Assigning values in x and y
```

```python
regression_funct(x,y) # Using regression function and check results.
```

I have used the golden ratio to train the machine learning model and test which is 70% Train and 30% to test.

The best result shows the XGBRegressor with 89% accuracy and RMSE of 2.3, MAE of 1.8 followd by GradientBoostingRegressor with 89% and ElasticNet with 73%. Therefore if we use XGBRegressor or GradientBoostingRegressor we will get most accurate results.

```python
# import Deep Learning libraries
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras.optimizers import Adam
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import  r2_score, mean_squared_error, mean_absolute_error
scaler=MinMaxScaler()
```

```python
x,y=Main.drop("MEDV",axis=1),Main[["MEDV"]] # Assign x and y
x=scaler.fit_transform(x) # scale x values
```

```python
# split data in test and train
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.30, random_state=42)
```

```python
# Build the Model
model=Sequential()
model.add(Dense(200, activation='relu'))
model.add(Dense(200, activation='relu'))
model.add(Dense(200, activation='relu'))
model.add(Dense(200, activation='relu'))
model.add(Dense(200, activation='relu'))
model.add(Dense(200, activation='relu'))
model.add(Dense(200, activation='relu'))
model.add(Dense(200, activation='relu'))
model.add(Dense(200, activation='relu'))
model.add(Dense(1))
model.compile(optimizer='Adam', loss='mse')#mse regression
```

```python
model.fit(x=x_train, y=y_train, validation_data=(x_test, y_test), batch_size=128, epochs=500, verbose=1)
```
